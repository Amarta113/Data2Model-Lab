#Car Price Prediction

```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv("/content/car_price_dataset[1].csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

Check is there any null values

```python
df.isnull().sum()
```

```python
df.info()
```

Let's check categorical columns:

```python
categories = df.select_dtypes(include='object').columns.tolist()
```

```python
categories
```

Convert the categorical data to numeric form with one hot encoding method:

```python
encoded_col = pd.get_dummies(df[categories], drop_first=True)
```

let's check first five rows of encoded columns

```python
encoded_col.head()
```

Concatenate the encoded data with the actual dataframe

```python
df = pd.concat([df, encoded_col], axis=1)
```

```python
df.head()
```

Drop the categorical columns

```python
df.drop(categories, inplace=True, axis=1)
```

```python
df.head()
```

```python
df.info()
```

Split the features to x and target variable to y:

```python
x = df.drop('Price', axis=1)
y = df['Price']
```

```python
y.shape
```

Let's perform normalization with Standard Scalar

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_transform = scaler.fit_transform(x)
```

Split the training data and testing data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_transform, y, test_size=.20, random_state=101)
```

Perform linear Regression:

```python
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(x_train, y_train)
```

```python
y_predicted = linear_regressor.predict(x_test)
```

```python
y_test.shape
```

Differentiate the predicted values and actual values:

```python
results = np.column_stack((y_test, y_predicted))

print("Actual values | Predicted Values \n"+"--"*20)
for actual, predicted in results:
  print(f"{actual: 14.2f} | {predicted:14.2f}")
```

Evaluate the model prediction

```python
from sklearn.metrics import mean_squared_error, r2_score

accuracy = r2_score(y_test, y_predicted)*100
mse = mean_squared_error(y_test, y_predicted)
rmse = np.sqrt(mse)

print("Mean Squared Error: ", mse)
print("Root Mean Squared Error: ", rmse)
print("Accuracy: ", accuracy)
```

Decision Tree Regression:

```python
from sklearn.tree import DecisionTreeRegressor

DT_regressor = DecisionTreeRegressor()
DT_regressor.fit(x_train, y_train)

dt_reg_predicted = DT_regressor.predict(x_test)

dt_reg_mse = mean_squared_error(y_test, dt_reg_predicted)
dt_reg_accuracy = r2_score(y_test, dt_reg_predicted)*100
dt_reg_rmse = np.sqrt(dt_reg_mse)

print("Decion Tree Regressor : \n"+"--"*30)
print("Mean Squared Error: ", dt_reg_mse)
print("Root Mean Squared Error: ", dt_reg_mse)
print("Accurcay: ", dt_reg_accuracy)
```

Random Forest Regression:

```python
from sklearn.ensemble import RandomForestRegressor

RF_regressor = RandomForestRegressor()
RF_regressor.fit(x_train, y_train)

rf_reg_predicted = RF_regressor.predict(x_test)

rf_reg_mse = mean_squared_error(y_test, rf_reg_predicted)
rf_reg_rmse = np.sqrt(rf_reg_mse)
rf_reg_accuracy = r2_score(y_test, rf_reg_predicted)*100

print("Random Forest Regressor : \n"+"--"*30)
print("Mean Squared Error: ", rf_reg_mse)
print("Root Mean Squared Error: ", rf_reg_mse)
print("Accurcay: ", rf_reg_accuracy)
```

Gradient Boosting Regression:

```python
from sklearn.ensemble import GradientBoostingRegressor

GB_regressor = GradientBoostingRegressor()
GB_regressor.fit(x_train, y_train)

gb_reg_predicted = GB_regressor.predict(x_test)

gb_reg_mse = mean_squared_error(y_test, gb_reg_predicted)
gb_reg_rmse = np.sqrt(gb_reg_mse)
gb_reg_accuracy = r2_score(y_test, gb_reg_predicted)*100

print("Gradient Boosting Regressor: \n"+"--"*30)
print("Mean Squared Error: ", gb_reg_mse)
print("Root Mean Squared Error: ", gb_reg_rmse)
print("Accuracy: ", gb_reg_accuracy)
```

Ranking the model performance

```python
model_score = {
    "Linear Regression" : accuracy,
    "Decision Tree Regression" : dt_reg_accuracy,
    "Random Forest Regression": rf_reg_accuracy,
    "Gradient Boosting Regression": gb_reg_accuracy
}

sorted_score = sorted(model_score.items(), key=lambda x: x[1], reverse=True)

print("Ranking of the model performance: ")
for rank, (model_name, score) in enumerate(sorted_score, start=1):
  print(f"{rank}. {model_name} : {score}")
```
